```python
from tkinter.font import Font
import numpy as np
import matplotlib.pyplot as plt

from comparisons.transfer_study import E_SC

# ---- raw table (include BE1D, we'll drop it) ----
#            name      NBG-PT      NBG-SC     GCR
raw = [
    ("CFD1D",        0.1454689984, 0.5818485829, 0.2500117775),
    ("CFD2D",        0.06192615745, 0.4805626099, 0.1288617886),
    ("CFD3D",        0.06680265171, 0.3741075982, 0.1785653433),
    ("MHD3D",    0.06872083432, 0.6441928522, 0.1066774244),
    ("CFD3D-T",   0.06690049669, 0.8046891552, 0.08313831032),
    ("TGC3D",    0.07272096063, 0.665781735,  0.1092264278),
]

# ---- drop BE1D and split columns ----
rows = [r for r in raw]
labels = [r[0] for r in rows]
NBG_PT   = [r[1] for r in rows]
NBG_SC   = [r[2] for r in rows] 
GCR    = [r[3] for r in rows]

# ---- grouped bar chart with specific colors + savefig ----
x = np.arange(len(labels))
w = 0.2
colors = ['tab:blue', 'tab:green', 'tab:orange']

fig, ax = plt.subplots(figsize=(8, 3.5))

ax.bar(x - 1.5*w, NBG_PT, width=w, label="NBG-PT", color=colors[0])
ax.bar(x - 0.5*w, GCR, width=w, label="GCR", color=colors[1])
ax.bar(x + 0.5*w, NBG_SC, width=w, label="NBG-SC", color=colors[2])

ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=20, ha="right")
ax.set_ylabel("Value")
ax.legend(loc="best")
fig.tight_layout()

# save as PNG at 300 dpi
fig.savefig("transfer_study_bar.png", dpi=500, bbox_inches="tight")

plt.show()
```
